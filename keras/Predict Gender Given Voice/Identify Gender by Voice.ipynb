```python
import pandas as pd
import numpy as np
from keras import backend as K
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
url = "voice.csv"
data = pd.read_csv(url)
data.shape
```

```python
data['label'].replace("female", 0,inplace=True)
data['label'].replace("male",1,inplace=True)
data.head()
```

```python
train, test = train_test_split(data,test_size = 0.3) 
```

```python
gender_train = train["label"]
gender_test = test["label"]
voice_train = train.drop("label",1)
voice_test = test.drop("label",1)
```

```python
clf = MLPClassifier(hidden_layer_sizes = (30,25,10),activation = "tanh", max_iter = 1000)
clf.fit(voice_train,gender_train)
```

```python
clf.score(voice_test,gender_test)
```

```python
t_clf = RandomForestClassifier()
t_clf.fit(voice_train,gender_train)
```

```python
t_clf.score(voice_test,gender_test)
```

```python
v_train = voice_train.values
g_train = gender_train.values
v_test = voice_test.values
g_test = gender_test.values
Y_train = np_utils.to_categorical(g_train,2)
Y_test = np_utils.to_categorical(g_test,2)
model = Sequential()
model.add(Dense(35,activation = "tanh", input_shape = (20,)))
model.add(Dense(20,activation = "tanh"))
model.add(Dropout(0.3))
model.add(Dense(5,activation = "sigmoid"))
model.add(Dense(2,activation = "sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
model.fit(v_train,Y_train, batch_size = 50, epochs = 200, verbose = 1)
score = model.evaluate(v_test, Y_test, verbose=0)
print(score)
```

```python
amount_wrong = 0
for x in range(0,len(g_test)):
    prediction = np.argmax(model.predict(v_test)[x])
    if g_test[x] != prediction:
        amount_wrong = amount_wrong + 1
    #print([prediction,g_test[x]])
print(amount_wrong)
```

```python
model.save("voice_model.h5")
```
